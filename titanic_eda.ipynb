```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("data/Titanic-Dataset.csv")
```

```python
df['Embarked'].mode()
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace= True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace= True)
```

```python
df.drop(columns=['Cabin'], inplace= True)
```

```python
sns.histplot(df['Age'], kde= True)
plt.title ('Age distribution')
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title ('Gender  distribution')
plt.show()
```

```python
sns.boxplot(x=df['Fare'])
plt.title('Fare boxplot')
plt.show()
```

```python
sns.countplot(x= 'Sex', hue='Survived', data=df)
plt.title ("Gender vs Survival")
plt.show()
```

```python
sns.countplot(x= "Pclass", hue= "Survived", data=df )
plt.title("Passenger class vs Survival")
plt.show()
```

```python
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs survival")
plt.show()
```

```python
df['Age-group']= pd.cut(df['Age'], bins=[0,12,18,35,60,100], labels= ["Child", "Teenagers", "Young Adults", "Adult", "Senior"])
```

```python
sns.countplot (x= 'Age-group', hue= "Survived", data=df)
plt.title("Age vs Survival")
plt.show()
```

#  Final Insights

1. Female passengers had higher survival rates than male passengers.

2. First-class passengers were more likely to survive compared to lower-class passengers.

3. Children showed relatively better survival proportions than many adult passengers.

4. Most passengers were young adults and adults.

5. The dataset contained missing values and fare outliers, which were handled during preprocessing.
